Importing libraries required to read the csv file of the dataset. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/dataset1/master/mushrooms.csv')
df
```

```python
df.isnull().sum()
```

Dataset doesn't have any null values. 

Coverting the dataset to numerical values using encoder. 

```python
sns.countplot(df['class'])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_new = df.apply(le.fit_transform)
df_new
```

We have now coverted the string inputs to numerical and we can check that the Class Column(Labels are coverted to 0 and 1) where 0 is poisnous and 1 is Edible one.

```python
x = df_new.iloc[:,1:23]
x
```

```python
y = df_new['class']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=30,test_size=.30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
predict = dt.predict(x_test)
```

```python
print(accuracy_score(y_test,predict))
```

```python
print(confusion_matrix(y_test,predict))
```

```python
print(classification_report(y_test,predict))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
predict_lg = lg.predict(x_test)
```

```python
print(accuracy_score(y_test,predict_lg))
```

```python
print(confusion_matrix(y_test,predict_lg))
```

```python
print(classification_report(y_test,predict_lg))
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive = MultinomialNB()
```

```python
naive.fit(x_train,y_train)
```

```python
predict_naive = naive.predict(x_test)
```

```python
print(accuracy_score(y_test,predict_naive))
```

```python
print(confusion_matrix(y_test,predict_naive))
```

```python
print(classification_report(y_test,predict_naive))
```

```python
from sklearn.svm import SVC
```

```python
sup_vect = SVC()
```

```python
sup_vect.fit(x_train,y_train)
```

```python
predict_sup_vect = sup_vect.predict(x_test)
```

```python
print(accuracy_score(y_test,predict_sup_vect))
print('\n')
print(confusion_matrix(y_test,predict_sup_vect))
print('\n')
print(classification_report(y_test,predict_sup_vect))
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel = 'linear', degree = 5 , max_iter = 2)

svm.fit(x_train,y_train)

predict_svm = svm.predict(x_test)

print(accuracy_score(y_test,predict_svm))
print('\n')
print(confusion_matrix(y_test,predict_svm))
print('\n')
print(classification_report(y_test,predict_svm))
```
